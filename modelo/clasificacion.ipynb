```python
import pandas as pd
import shutil
import os
from concurrent.futures import ThreadPoolExecutor
import random
```

```python
csv_file = 'train.csv'  # Reemplaza con la ruta a tu archivo CSV
images_dir = 'train_images'  # Reemplaza con la ruta a la carpeta que contiene las imágenes
output_dir = 'new_images'  # Reemplaza con la ruta a la carpeta donde quieres mover las imágenes sanas
```

```python
os.makedirs(output_dir, exist_ok=True)

# Cargar el archivo CSV con pandas
df = pd.read_csv(csv_file)

# Filtrar las imágenes con label = 4 (sanas)
sanas = df[df['label'] == 4]

def move_image(image_name):
    """Función para mover una imagen de la clase sana a la carpeta de salida."""
    image_path = os.path.join(images_dir, image_name)
    
    if os.path.exists(image_path):
        shutil.copy(image_path, os.path.join(output_dir, image_name))
        print(f"Imagen {image_name} movida a la carpeta sana.")
    else:
        print(f"Imagen {image_name} no encontrada en la carpeta de imágenes.")

with ThreadPoolExecutor(max_workers=8) as executor:  
    executor.map(move_image, sanas['image_id'])

print("Proceso completado.")
```

```python
sana_dir = 'C:/Users/juan0/Desktop/Backend-Tesis/modelo/data/sana' 
images_in_sana = os.listdir(sana_dir)

all_images = os.listdir(output_dir)  # Listar todas las imágenes en la carpeta sana

images_to_move = [image for image in all_images if image not in images_in_sana]

if len(images_to_move) > 150:
    images_to_keep = random.sample(images_to_move, 150)  # Seleccionar aleatoriamente 150 imágenes
else:
    images_to_keep = images_to_move  # Si hay menos de 150, seleccionar todas


images_to_delete = [image for image in all_images if image not in images_to_keep]

for image in images_to_delete:
    image_path = os.path.join(output_dir, image)
    os.remove(image_path)  # Eliminar la imagen
    print(f"Imagen {image} eliminada.")

print("Proceso de eliminación de imágenes completado.")
```
